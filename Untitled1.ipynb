```python
df['date'] = pd.to_datetime(df['date'])
df['Day'] = df['date'].dt.strftime('%d')
df['month'] = df['date'].dt.strftime('%m')
df['year'] = df['date'].dt.strftime('%Y')
df.drop(['date'],axis=1,inplace=True)
df.drop(['street'],axis = 1,inplace=True)
df.drop(['city'],axis = 1,inplace=True)
df.drop(['statezip'],axis = 1,inplace=True)
df.drop(['country'],axis = 1,inplace=True)
df.head()
```

## Creating a Pipeline 

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy="most_frequent")),
    ('std_scaler',StandardScaler())
])
```

```python
housing_tr = my_pipeline.fit_transform(df)
housing_tr
housing_tr.shape
```

```python
y=df['price']
X=df.drop(['price'],axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X)

X2=scaler.transform(X)
```

## Train Test Splitting 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X2, y, test_size=0.33, random_state=42)
```

## Evaluating Models 

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import f1_score
def accuracy(model):
    model.fit(X_train, y_train)
    predict = model.predict(X_test)
    return y_test, predict
```

```python
# from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import SGDRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn import svm
from sklearn.cluster import KMeans
from sklearn.ensemble import AdaBoostRegressor


# model1 = XGBRegressor()
model2 = RandomForestRegressor()
#model3 = SVC()
model4 = LogisticRegression()
model5 = KNeighborsRegressor()
model6 = SGDRegressor()
model7 = DecisionTreeRegressor()
model8 = GaussianNB()
model9 = SVC()
model10= AdaBoostRegressor()
#model11=AdaBoostClassifier()
    
models=[model2,model5,model6,model7,model10]
i=0


for model in models:
    i+=1
    y_test, predict=accuracy(model)
    print("Model ", i,":", model)
    print('mean_squared_error score:',mean_squared_error(y_test, predict))
#     print('F1 score:',f1_score(y_test, predict))
```
